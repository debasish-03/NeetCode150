## [Valid Anagram](https://leetcode.com/problems/valid-anagram/description/)

```python
Given two strings s and t, return true if t is an anagram of s, and false otherwise.

 

Example 1:

Input: s = "anagram", t = "nagaram"

Output: true

Example 2:

Input: s = "rat", t = "car"

Output: false

 

Constraints:

1 <= s.length, t.length <= 5 * 104
s and t consist of lowercase English letters.
 

Follow up: What if the inputs contain Unicode characters? How would you adapt your solution to such a case?
```

### Python Solution

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        count_s = dict()
        for ch in s:
            count_s[ch] = 1 + count_s.get(ch, 0)

        for ch in t:
            if ch not in count_s:
                return False

            count_s[ch] -= 1
       
        return all(val == 0 for val in count_s.values())


Time Complexity: O(n)
Space Complexity: O(1) - Since the problem states that the strings consist of lowercase English letters, the maximum number of unique characters is 26.

```

### Java Solution - 1

class Solution {
    public boolean isAnagram(String s, String t) {
        if (s.length() != t.length()) {
            return false;
        }

        Map<Character, Integer> mapOfS = new HashMap<>();
        for (int i = 0; i < s.length(); i++) {
            mapOfS.put(s.charAt(i), 1 + mapOfS.getOrDefault(s.charAt(i), 0));
        }

        for (int i = 0; i < t.length(); i++) {
            if (!mapOfS.containsKey(t.charAt(i))) {
                return false;
            }
            mapOfS.put(t.charAt(i), mapOfS.get(t.charAt(i)) - 1);
        }

        for (Integer value: mapOfS.values()) {
            if (value != 0) {
                return false;
            }
        }

        return true;
    }
        
}

Time Complexity: O(n)
Space Complexity: O(1) - Since the problem states that the strings consist of lowercase English letters, the maximum number of unique characters is 26.
