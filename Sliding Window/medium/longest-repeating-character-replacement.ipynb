# [Longest Repeating Character Replacement](https://leetcode.com/problems/longest-repeating-character-replacement/)
```python
You are given a string s and an integer k. You can choose any character of the string and change it to any other uppercase English character. You can perform this operation at most k times.

Return the length of the longest substring containing the same letter you can get after performing the above operations.

 

Example 1:

Input: s = "ABAB", k = 2
Output: 4
Explanation: Replace the two 'A's with two 'B's or vice versa.
Example 2:

Input: s = "AABABBA", k = 1
Output: 4
Explanation: Replace the one 'A' in the middle with 'B' and form "AABBBBA".
The substring "BBBB" has the longest repeating letters, which is 4.
There may exists other ways to achieve this answer too.
 

Constraints:

1 <= s.length <= 105
s consists of only uppercase English letters.
0 <= k <= s.length
```

### Python Solution
```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        freq = {}
        left, right = 0, 0
        res = 0
        maxi = 0
        while right < len(s):
            freq[s[right]] = 1 + freq.get(s[right], 0)
            maxi = max(maxi, freq[s[right]])

            if (right - left + 1) - maxi > k:
                freq[s[left]] -= 1
                left += 1

            res = max(res, right - left + 1)
            right += 1

        return res

        
       
Time Complexity: O(N), where N is the length of the string s.
Space Complexity: O(1), since the frequency dictionary will hold at most 26 entries for each uppercase English letter.
```

### Java Solution
```java
class Solution {
    public int characterReplacement(String s, int k) {
        int res = 0;
        int left = 0;
        int right = 0;
        int maxi = 0;

        Map<Character, Integer> freq = new HashMap<>();
        while (right < s.length()) {
            char ch = s.charAt(right);
            freq.put(ch, 1 + freq.getOrDefault(ch, 0));
            maxi = Math.max(maxi, freq.get(ch));

            if ((right - left + 1) - maxi > k) {
                freq.put(s.charAt(left), freq.get(s.charAt(left)) - 1);
                left++;

            }

            res = Math.max(res, (right - left + 1));
            right++;
        }

        return res;
    }
}


Time Complexity: O(N), where N is the length of the string s.
Space Complexity: O(1), since the frequency map will hold at most 26 entries for each uppercase English letter.
```
