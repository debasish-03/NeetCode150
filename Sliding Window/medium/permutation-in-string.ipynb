# [Permutation in String](https://leetcode.com/problems/permutation-in-string/description/)

```python
Given two strings s1 and s2, return true if s2 contains a permutation of s1, or false otherwise.

In other words, return true if one of s1's permutations is the substring of s2.

 

Example 1:

Input: s1 = "ab", s2 = "eidbaooo"
Output: true
Explanation: s2 contains one permutation of s1 ("ba").
Example 2:

Input: s1 = "ab", s2 = "eidboaoo"
Output: false
 

Constraints:

1 <= s1.length, s2.length <= 104
s1 and s2 consist of lowercase English letters.
```

### Python Solution 
```python
class Solution:
    def checkInclusion(self, s1: str, s2: str) -> bool:
        if len(s1) > len(s2):
            return False
        
        permutation_map = {}
        for ch in s1:
            permutation_map[ch] = 1 + permutation_map.get(ch, 0)
        
        window = {}
        for ch in s2[:len(s1)]:
            window[ch] = 1 + window.get(ch, 0)

        if window == permutation_map:
            return True

        for i in range(len(s1), len(s2)):
            left_char = s2[i - len(s1)]
            right_char = s2[i]

            window[left_char] -= 1
            if window[left_char] == 0:
                del window[left_char]

            window[right_char] = 1 + window.get(right_char, 0)
            if window == permutation_map:
                return True


        return False


        
Time Complexity: O(N) where N is the length of s2.
Space Complexity: O(1) since the size of the hashmap is bounded by the number of lowercase English letters (26).
```

## Java Solution
```java
class Solution {
    public boolean checkInclusion(String s1, String s2) {
        if (s1.length() > s2.length()) return false;
        
        Map<Character, Integer> perMap = new HashMap<>();
        for (int i = 0; i < s1.length(); i++) {
            char ch = s1.charAt(i);
            perMap.put(ch, 1 + perMap.getOrDefault(ch, 0));
        }

        Map<Character, Integer> window = new HashMap<>();
        for (int i = 0; i < s1.length(); i++) {
            char ch = s2.charAt(i);
            window.put(ch, 1 + window.getOrDefault(ch, 0));
        }

        if (window.equals(perMap)) {
            return true;
        }

        for (int i = s1.length(); i < s2.length(); i++) {
            char left = s2.charAt(i - s1.length());
            char right = s2.charAt(i);

            window.put(left, window.get(left) - 1);
            if (window.get(left) == 0) {
                window.remove(left);
            }

            window.put(right, 1 + window.getOrDefault(right, 0));
            if (window.equals(perMap)) {
                return true;
            }
        }

        return false;
    }
}



Time Complexity: O(N) where N is the length of s2.
Space Complexity: O(1) since the size of the hashmap is bounded by the number of lowercase English letters (26).
```
