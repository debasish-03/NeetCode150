# [Trapping Rain Water](https://leetcode.com/problems/trapping-rain-water/description/)
```python
Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

 
Example 1:
```

```python
Example 2:

Input: height = [4,2,0,3,2,5]
Output: 9
 

Constraints:

n == height.length
1 <= n <= 2 * 104
0 <= height[i] <= 105
```

### Python Solution
```python
class Solution:
    def trap(self, h: List[int]) -> int:
        water = 0
        l_max = 0
        r_max = 0

        left = 0
        right = len(h) - 1

        while left < right:
            if h[left] <= h[right]:
                if l_max < h[left]:
                    l_max = h[left]
                else:
                    water += l_max - h[left]
                left += 1
            else:
                if r_max < h[right]:
                    r_max = h[right]
                else:
                    water += r_max - h[right]
                right -= 1

        return water
            

Time Complexity: O(n)
Space Complexity: O(1)
```

## Java Solution
```java
class Solution {
    public int trap(int[] h) {
        int water = 0;
        int l = 0;
        int r = h.length - 1;

        int lMax = 0;
        int rMax = 0;

        while (l < r) {
            if (h[l] <= h[r]) {
                if (lMax < h[l]) {
                    lMax = h[l];
                } else {
                    water += lMax - h[l];
                }
                l++;
            } else {
                if (rMax < h[r]) {
                    rMax = h[r];
                } else {
                    water += rMax - h[r];
                }
                r--;
            }
        }
        return water;
    }
}

Time Complexity: O(n)
Space Complexity: O(1)
```
